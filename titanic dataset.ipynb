```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Milan\Downloads\train (1).csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(x='Pclass', y = 'Age', data = df, palette='viridis_r')
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24
    else :
        return Age
```

```python
df.isnull().sum()
```

```python
df.drop('Cabin', axis=1, inplace=True)
df
```

```python
df['Age'] = df[['Age', 'Pclass']].apply(impute_age, axis =1)
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cmap='viridis')
```

```python
df.isnull().sum()
```
